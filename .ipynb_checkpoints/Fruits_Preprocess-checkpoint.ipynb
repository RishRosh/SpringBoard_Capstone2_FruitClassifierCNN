## Pre-process images

In this notebook, the images are pre-processed by thumbnailing, padding them with white pixels and either conversion to grayscale or keeping the colors. Attempts are made by using multiple libraries of Pillow and CV2 and comparing the classifier accuracy. It was found that colored images preproceesed by CV2 have better accuracy.

<b>Note:</b> Classifier training has been done on COLAB notebooks to take advantage of free GPU available. Appropriate references to original authors/inspiration have been given in respective sections. Lots of improvisation has been done.

<b>Ref1:</b> Pillow library techniques for image processing taken from: https://machinelearningmastery.com/how-to-load-and-manipulate-images-for-deep-learning-in-python-with-pil-pillow/<br>
<b>Ref2:</b> All CV2 image processing techniques are inspired by: https://jdhao.github.io/2017/11/06/resize-image-to-square-with-padding/ <br>
<b>Ref3:</b> Keras processing technique inspired by : https://towardsdatascience.com/classify-butterfly-images-with-deep-learning-in-keras-b3101fe0f98

```python
def make_thumb_gray_pad(im, to_size=100, fill_color=(0,0,0,0)):
    """ Function in Pillow to resize, padd image and convert to grayscale"""
    x, y = im.size
    size = to_size
    # Pillow doesn't cater for smaller images than desired size. This is for accomodating that and scaling image to highest side.
    size = max(to_size, x,y) 
    new_im = Image.new('L', (size, size), (255)) # Create a blank grayscale image with white background
#     new_im = Image.new('RGBA', (size, size), fill_color) # Use this for color images
    new_im.paste(im, (int((size-x)/2), int((size - y)/2))) # Paste in to the blank image by centering the original
    new_im_rs = new_im.resize((to_size, to_size), Image.ANTIALIAS) # resize to thumbnail
    return new_im_rs

def make_thumb_pad(im, to_size=100, fill_color=(0,0,0,0)):
    """ Function in Pillow to resize, padd image and convert to grayscale"""
    x, y = im.size
    size = to_size
    # Pillow doesn't cater for smaller images than desired size. This is for accomodating that and scaling image to highest side.
    size = max(to_size, x,y) 
#     new_im = Image.new('L', (size, size), (255)) # Create a blank grayscale image with white background
    new_im = Image.new('RGBA', (size, size), fill_color) # Use this for color images
    new_im.paste(im, (int((size-x)/2), int((size - y)/2))) # Paste in to the blank image by centering the original
    new_im_rs = new_im.resize((to_size, to_size), Image.ANTIALIAS) # resize to thumbnail
    return new_im_rs


# Ref: https://stackoverflow.com/questions/44231209/resize-rectangular-image-to-square-keeping-ratio-and-fill-background-with-black/44231784   
```

```python
def cv2_resize_gray_pad(im, desired_size = 100, border_color='w'):
    """ Function to resize, gray and pad image in CV2 """
    old_size = im.shape[:2] # old_size is in (height, width) format
    
    # First resize image
    ratio = float(desired_size)/max(old_size)
    new_size = tuple([int(x*ratio) for x in old_size])
    im = cv2.resize(im, (new_size[1], new_size[0]), interpolation = cv2.INTER_AREA)

    # Now pad image by using border to fill it
    delta_w = desired_size - new_size[1]
    delta_h = desired_size - new_size[0]
    top, bottom = delta_h//2, delta_h-(delta_h//2)
    left, right = delta_w//2, delta_w-(delta_w//2)

    if border_color =='w':
        color = [255, 255, 255]
    if border_color =='b':
        color = [0, 0, 0]
        
    new_im = cv2.copyMakeBorder(im, top, bottom, left, right, cv2.BORDER_CONSTANT, value=color)
    new_im_gray = cv2.cvtColor(new_im, cv2.COLOR_BGR2GRAY) # Convert to grayscale
    return new_im_gray

def cv2_resize_pad(im, desired_size = 100, border_color='w'):
    """ Function to resize, gray and pad image in CV2 """
    old_size = im.shape[:2] # old_size is in (height, width) format
    
    # First resize image
    ratio = float(desired_size)/max(old_size)
    new_size = tuple([int(x*ratio) for x in old_size])
    im = cv2.resize(im, (new_size[1], new_size[0]), interpolation = cv2.INTER_AREA )

    # Now pad image by using border to fill it
    delta_w = desired_size - new_size[1]
    delta_h = desired_size - new_size[0]
    top, bottom = delta_h//2, delta_h-(delta_h//2)
    left, right = delta_w//2, delta_w-(delta_w//2)

    if border_color =='w':
        color = [255, 255, 255]
    if border_color =='b':
        color = [0, 0, 0]
        
    new_im = cv2.copyMakeBorder(im, top, bottom, left, right, cv2.BORDER_CONSTANT, value=color)
#     new_im_gray = cv2.cvtColor(new_im, cv2.COLOR_BGR2GRAY) # Convert to grayscale
    return new_im
    
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(df, test_size=0.2, random_state=34)

datagen=ImageDataGenerator(
    rotation_range = 40,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    rescale=1./255)

train_generator=datagen.flow_from_dataframe(
    dataframe=train_set, 
    directory="H:\\DataScience\\Cap2\\train_set_col", 
    x_col="id", 
    y_col="label", 
    class_mode="categorical", 
    target_size=(100,100), 
    batch_size=32,
    save_to_dir='H:\\DataScience\\Cap2\\tempop',
    shuffle=True)

valid_generator=datagen.flow_from_dataframe(
    dataframe=test_set, 
    directory="H:\\DataScience\\Cap2\\train_set_col", 
    x_col="id", 
    y_col="label", 
    class_mode="categorical", 
    target_size=(100,100), 
    batch_size=32,
    save_to_dir='H:\\DataScience\\Cap2\\tempop',
    shuffle=True)
```
