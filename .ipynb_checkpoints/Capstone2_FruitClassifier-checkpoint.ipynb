# Capstone 2: Fruit Classifier

This notebook details all the steps that were carried out during the making of fruit classifier project, which got trained on color images on Google's COLAB environment. <br>

### Objective
The objective is to train a CNN (Convoluted Neural Network) to recognize everyday fruits in at least 5 categories. This would be an end to end activity which will involve data curation, data transformation, augmentation, training, validation and testing to achieve at least 90% accuracy on test set (unseen data). They will be trained using 300 color images of each category that are collected from free online resources. For testing, 10 images will be taken for each class. There will be five categories of fruits Red Apples, Strawberries, Green Grapes, Kiwi and Blackberries. The objective is to build a foundational level classifier that can be scaled up to applications in numerous technologies like facial recognition, computer vision, self-driving cars etc. The CNN will need to ensure that it gets over all the below challenges in achieving its target accuracy.

<b>Challenge 1:</b> Color similarity – There will complexity of having at least two sets of fruits with same color. This will up the bar by minimizing any color related differentiation by the CNN. For example in this set, strawberries and apples will be of red color and grapes and kiwis will be of green color. This will ensure that the CNN doesn’t discriminate largely based on color of fruits. It could be the case if these 5 fruits were chosen of very different colors.<br> 
<b>Challenge 2:</b> Variable Shape – The standard variations in the fruits’ shape should be accounted for. For example rounder grapes vs. more oval variation. Differences in shapes of strawberries, apples etc.<br>
<b>Challenge 3:</b> Foreign objects/ partial object recognition – Images that are partial populated with the fruit or have foreign objects blocking it should also be recognizable.

```python
# Define a function to download the images that takes in keywords, size and number of images to downlaod

from flickrapi import FlickrAPI  # https://pypi.python.org/pypi/flickrapi
import urllib
import os
from random import randint
import time

def download_flickr_photos(keywords, size='original', max_nb_img=-1):
    """
    Downloads images based on keyword search on the Flickr website
    
    Parameters
    ----------
    keywords : string, list of strings
        Keyword to search for or a list of keywords should be given.
    size : one of the following strings 'thumbnail', 'square', 'medium', default: 'original'.
        Size of the image to download. In this function we only provide
        four options. More options are explained at 
        http://librdf.org/flickcurl/api/flickcurl-searching-search-extras.html
    max_nb_img : int, default: -1
        Maximum number of images per keyword to download. If given a value of -1, all images
        will be downloaded
    
    Returns
    ------
    Images found based on the keyword are saved in a separate subfolder.
    
    Notes
    -----
    This function uses the Python package flickrapi and its walk method. 
    FlickrAPI.walk has same parameters as FlickrAPI.search
    http://www.flickr.com/services/api/flickr.photos.search.html
    
    To use the Flickr API a set of API keys needs to be created on 
    https://www.flickr.com/services/api/misc.api_keys.html
    """
    if not (isinstance(keywords, str) or isinstance(keywords, list)):
        raise AttributeError('keywords must be a string or a list of strings')
        
    if not (size in ['thumbnail', 'square', 'medium', 'original']):
        raise AttributeError('size must be "thumbnail", "square", "medium" or "original"')
                             
    if not (max_nb_img == -1 or (max_nb_img > 0 and isinstance(max_nb_img, int))):
        raise AttributeError('max_nb_img must be an integer greater than zero or equal to -1')
    
    flickr = FlickrAPI(API_KEY, API_SECRET)
    
    if isinstance(keywords, str):
        keywords_list = []
        keywords_list.append(keywords)
    else:
        keywords_list = keywords
        
    if size == 'thumbnail':
        size_url = 'url_t'
    elif size == 'square':
        size_url = 'url_q'
    elif size == 'medium':
        size_url = 'url_c'
    elif size == 'original':
        size_url = 'url_o'
    
    for keyword in keywords_list:
        count = 0
                             
        #print('Downloading images for', keyword)

        results_folder = IMG_FOLDER + keyword.replace(" ", "_") + "/"
        if not os.path.exists(results_folder):
            os.makedirs(results_folder)

        photos = flickr.walk(
                     text=keyword,
                     extras=size_url,
                     license='1,2,4,5',
                     sort='relevance', # Added improvisation
                     per_page=50)
        
        urls = []
        for photo in photos:
            t = randint(1, 3)
            time.sleep(t)
            count += 1
            if max_nb_img != -1:
                if count > max_nb_img:
                    print('Reached maximum number of images to download')
                    break
            try:
                url=photo.get(size_url)
                urls.append(url)
                
                urllib.request.urlretrieve(url,  results_folder + str(count) +".jpg")
                print('Downloading image #' + str(count) + ' from url ' + url)
            except Exception as e:
                print(e, 'Download failure')
                             
        print("Total images downloaded:", str(count - 1))
```

```python
# import the necessary packages
from imutils import paths
import argparse
import requests
import cv2
import os


# construct the argument parse and parse the arguments
ap = argparse.ArgumentParser()
ap.add_argument(
    "-u",
    "--urls", 
    help="H:\\DataScience\\Cap2\\")

ap.add_argument(
    "-o",
    "--output",
    help="H:\\DataScience\\Cap2\\Green_Grapes\\")

```

```python
# Define the function to download images

def download_imgs(arg_str):
    args = ap.parse_args(arg_str.split())
    rows = open(args.urls).read().strip().split("\n")
    total = 0
    
    # loop the URLs
    for url in rows:
        try:
            # try to download the image
            r = requests.get(url, timeout=60)

            # save the image to disk
            p = os.path.sep.join([args.output, "{}.jpg".format(str(total).zfill(8))])
            f = open(p, "wb")
            f.write(r.content)
            f.close()

            # update the counter
            print("[INFO] downloaded: {}".format(p))
            total += 1

        # handle if any exceptions are thrown during the download process
        except:
            print("[INFO] error downloading {}...skipping".format(p))
    
    
    # loop over the image paths we just downloaded to cull bad images
    for imagePath in paths.list_images(args.output):
        # initialize if the image should be deleted or not
        delete = False

        # try to load the image
        try:
            image = cv2.imread(imagePath)

            # if the image is `None` then we could not properly load it
            # from disk, so delete it
            if image is None:
                delete = True

        # if OpenCV cannot load the image then the image is likely
        # corrupt so we should delete it
        except:
            print("Except")
            delete = True

        # check to see if the image should be deleted
        if delete:
            print("[INFO] deleting {}".format(imagePath))
            os.remove(imagePath)


```

### Step 2: Data Cleaning

##### Method 1: Auto-deletetion via OpenCV

In this, the function 'dowload_imgs' is employed to automatically delete the images that are bad.Bad images are classified as the ones that cannot be loaded by OpenCV's CV2. This is factored in to the function itself and the 'culling' part is run after all images are downloaded. From the execution log, you can clearly see which images were deleted.

##### Method 2: Manual

For purposes of data curation, each image was manually looked up and kept only if it fit the following rules:

1. Image is clear and is relevant representation of the object. Any non relevant images were deleted.
2. Image is of whole fruits and not slices 
<br>Note: This is because fruits can be carved/sliced in different ways and we would need a bigger set than 300 images to train different representations of the fruit. Exception to this is kiwi as there are more images of sliced Kiwi than whole Kiwi.
3. Image does not have other fruits in the picture, especially the ones we are training the CNN on. This would lead to confusion and misclassification. 
4. Images have similar distance from which it is taken. 
<br>Note: This is to avoid exteme zoom (Macro) or big distances to mislead the CNN. To be able to get to this level of sophistication, you would need a bigger set and even then, it would be hard to classify extremes/outliers.


##### Build train-test set. 

Images will be combined from the two sets downloaded using Flickr and Google and only the top 300 images of each class that satisfy the above criteria will be taken for training set. This makes a total of 1500 images that will be used for training. Out of the remaining good images, 10 images of each class will be taken for testing purposes. So a total of 50 images will act as an unseen test set for evaluating performance of the classifier.

We can now write functions to center the images, resize them to 100 x 100 and fill them with white pixels where they don't fit (padding). 

There are 2 versions of these functions, one for color and other for GrayScale. We are going to try converting all images to grayscale and train a CNN for comparing which one is better.


```python
# Image transformation functions in Pillow

def make_thumb_gray_pad(im, to_size=100, fill_color=(0,0,0,0)):
    """ Function in Pillow to resize, padd image and convert to grayscale"""
    x, y = im.size
    size = to_size
    # Pillow doesn't cater for smaller images than desired size. This is for accomodating that and scaling image to highest side.
    size = max(to_size, x,y) 
    new_im = Image.new('L', (size, size), (255)) # Create a blank grayscale image with white background
#     new_im = Image.new('RGBA', (size, size), fill_color) # Use this for color images
    new_im.paste(im, (int((size-x)/2), int((size - y)/2))) # Paste in to the blank image by centering the original
    new_im_rs = new_im.resize((to_size, to_size), Image.ANTIALIAS) # resize to thumbnail
    return new_im_rs

def make_thumb_pad(im, to_size=100, fill_color=(0,0,0,0)):
    """ Function in Pillow to resize, padd color image"""
    x, y = im.size
    size = to_size
    # Pillow doesn't cater for smaller images than desired size. This is for accomodating that and scaling image to highest side.
    size = max(to_size, x,y) 
#     new_im = Image.new('L', (size, size), (255)) # Create a blank grayscale image with white background
    new_im = Image.new('RGBA', (size, size), fill_color) # Use this for color images
    new_im.paste(im, (int((size-x)/2), int((size - y)/2))) # Paste in to the blank image by centering the original
    new_im_rs = new_im.resize((to_size, to_size), Image.ANTIALIAS) # resize to thumbnail
    return new_im_rs

# Ref: https://stackoverflow.com/questions/44231209/resize-rectangular-image-to-square-keeping-ratio-and-fill-background-with-black/44231784   
```

```python
def cv2_resize_gray_pad(im, desired_size = 100, border_color='w'):
    """ Function to resize, gray and pad image in CV2 """
    old_size = im.shape[:2] # old_size is in (height, width) format
    
    # First resize image
    ratio = float(desired_size)/max(old_size)
    new_size = tuple([int(x*ratio) for x in old_size])
    im = cv2.resize(im, (new_size[1], new_size[0]), interpolation = cv2.INTER_AREA)

    # Now pad image by using border to fill it
    delta_w = desired_size - new_size[1]
    delta_h = desired_size - new_size[0]
    top, bottom = delta_h//2, delta_h-(delta_h//2)
    left, right = delta_w//2, delta_w-(delta_w//2)

    if border_color =='w':
        color = [255, 255, 255]
    if border_color =='b':
        color = [0, 0, 0]
        
    new_im = cv2.copyMakeBorder(im, top, bottom, left, right, cv2.BORDER_CONSTANT, value=color)
    new_im_gray = cv2.cvtColor(new_im, cv2.COLOR_BGR2GRAY) # Convert to grayscale
    return new_im_gray

def cv2_resize_pad(im, desired_size = 100, border_color='w'):
    """ Function to resize and pad color image in CV2 """
    old_size = im.shape[:2] # old_size is in (height, width) format
    
    # First resize image
    ratio = float(desired_size)/max(old_size)
    new_size = tuple([int(x*ratio) for x in old_size])
    im = cv2.resize(im, (new_size[1], new_size[0]), interpolation = cv2.INTER_AREA)

    # Now pad image by using border to fill it
    delta_w = desired_size - new_size[1]
    delta_h = desired_size - new_size[0]
    top, bottom = delta_h//2, delta_h-(delta_h//2)
    left, right = delta_w//2, delta_w-(delta_w//2)

    if border_color =='w':
        color = [255, 255, 255]
    if border_color =='b':
        color = [0, 0, 0]
        
    new_im = cv2.copyMakeBorder(im, top, bottom, left, right, cv2.BORDER_CONSTANT, value=color)
#     new_im_gray = cv2.cvtColor(new_im, cv2.COLOR_BGR2GRAY) # Convert to grayscale
    return new_im
    
```

### Preprocess all raw data

Now we pre-process all raw data to generate 300 processed images of each of the fruits. 
For this, we have 5 sub-folders in folder called 'Full_Set_Raw', which after processing will end up in 'Full_Set_Processed'.


```python
# Preprocess all images using CV2, also writing a copy in the 'comb_dir', 
# which will have all processed images together in a single location

raw_dir = root_dir + 'Full_Set_Raw/'          # Define the raw folder from where the images will be read
proc_dir = root_dir + 'Full_Set_Processed/'   # Define the processed folder where processed images will be stored
comb_dir = root_dir + 'Full_Set_Proc_Comb/'   # Define the location for combined folder where processed images will be saved.
fruit_csv = proc_dir + 'train.csv'            # Define the location for training csv file
cnt = 1
category = 0

header = ['id', 'label']

with open(fruit_csv, 'w', newline='') as csvf:
    writer = csv.writer(csvf, delimiter=',')
    # Write Header
    writer.writerow(i for i in header)
    
    for subfolder in listdir(raw_dir):
        subDirPathSrc = raw_dir + subfolder + '/'
        subDirPathDest = proc_dir + subfolder + '/'

        for file in listdir(subDirPathSrc):
            img = cv2.imread(subDirPathSrc + file)
            new_im = cv2_resize_pad(img)
            cv2.imwrite(subDirPathDest + str(cnt) + '.jpg', new_im)
            cv2.imwrite(comb_dir + str(cnt) + '.jpg', new_im)
            writer.writerow([cnt, category])    
            cnt+=1
            
        category+=1

# Copy file from the processed directory to the processed and combined directory for training on Colab.    
system('copy ' + fruit_csv + ' ' + comb_dir)
```

```python
# Now build the test data set

test_raw_dir = root_dir + 'Test_Set_Raw/'          # Location of raw images
test_proc_dir = root_dir + 'Test_Set_Processed/'   # Location of processed images
test_comb_dir = root_dir + 'Test_Set_Proc_Comb/'   # Location of combined processed images
test_fruit_csv = test_proc_dir + 'test.csv'        # Location of test csv file
  


category = 0                                       # Reset the category variable, keep the count as we need to reuse it

with open(test_fruit_csv, 'w', newline='') as csvf:
    writer = csv.writer(csvf, delimiter=',')
    # Write Header
    writer.writerow(j for j in header)
    
    for subfolder in listdir(test_raw_dir):
        subDirPathSrc = test_raw_dir + subfolder + '\\'
        subDirPathDest = test_proc_dir + subfolder + '\\'

        for file in listdir(subDirPathSrc):
            img = cv2.imread(subDirPathSrc + file)
            new_im = cv2_resize_pad(img)
            cv2.imwrite(subDirPathDest + str(cnt) + '.jpg', new_im)
            cv2.imwrite(test_comb_dir + str(cnt) + '.jpg', new_im)
            writer.writerow([cnt, category])    
            cnt+=1
            
        category+=1
    
# Copy file from the processed directory to the processed and combined directory for training on Colab.    
system('copy ' + test_fruit_csv + ' ' + test_comb_dir)
```

```python
path_comb = 'data/Full_Set_Proc_Comb/'
train = pd.read_csv('data/Full_Set_Proc_Comb/train.csv')
train_image = []
for i in tqdm(range(train.shape[0])):
    img = image.load_img(path_comb + train['id'][i].astype('str')+'.jpg', target_size=(100,100,3)) # Load the images
    img = image.img_to_array(img) # Convert to arrays
    img = img/255 # Normalize the values as they are in greyscale 2 ^ 8 = 256. If in color, 256*256*256?
    train_image.append(img) # Append the list
X = np.array(train_image)

print('The maximum number in this 2d array is: ',np.amax(X))

y=train['label'].values
y = to_categorical(y)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)

# model from CNN cheatsheet
model3 = Sequential()
model3.add(Conv2D(32, kernel_size=(3, 3),activation='relu',input_shape=(100,100,3), padding='same'))
model3.add(Conv2D(32, (3, 3), activation='relu'))
model3.add(MaxPooling2D(pool_size=(2, 2)))
model3.add(Dropout(0.25))
model3.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
model3.add(Conv2D(64, (3, 3), activation='relu'))
model3.add(MaxPooling2D(pool_size=(2, 2)))
model3.add(Dropout(0.25))
model3.add(Flatten())
model3.add(Dense(512, activation='relu'))
model3.add(Dropout(0.5))
model3.add(Dense(5, activation='softmax'))

model3.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])

model3.fit(X_train, y_train, epochs=60, validation_data=(X_test, y_test), verbose=2)
```

### DATA AUGMENTATION: KERAS Image Data Generator and Flow options

Keras library has a good functionality of generating different types of images from one image by performing operations on it like rotation, shear, zoom, flip etc. These operations introduce subtle changes in the original image that can then be fed to CNN as extra images for training. This is a very good way of generating additional images from a base image to factor in variations and make the CNN adept. 

Another concept that Keras provides is that of 'FLOW'. It has 3 different methods for this in Keras documentation:
1. flow - Images need to be in array to be fed to the generator.
2. flow_from_directory - Images need to be in directories with directory names as classes
3. flow_from_dataframe - Images can be in folder, but names and classes come from DataFrame.

Lets first try just the <b> Option 1 - Keras flow </b>

```python
# Lets define a function to get images via numerical numbers, with a starting point

import matplotlib.image as mpimg
from matplotlib import rcParams

def show_imgs(x, start_num=0, dr='/', ext='jpg'):
    
    """ This function displays images with numerical sequential numbers that have a
        starting point. The name of the directory can be passed to it."""
    
    rcParams['figure.figsize'] = 17,3
    plt.rcParams['axes.grid'] = False
    fig, ax = plt.subplots(1,x)
    if x == 1:
        img_A = mpimg.imread(dr + str(start_num) + '.' + ext)
        ax.imshow(img_A)
        ax.set_xticklabels([])
        ax.set_yticklabels([])
    else:
        for i in range (0,x):
            img_A = mpimg.imread(dr + str(start_num + i) + '.' + ext)
            ax[i].imshow(img_A)
            ax[i].set_xticklabels([])
            ax[i].set_yticklabels([])
            plt.tight_layout()

# Example usage:
show_imgs(2, start_num=1501, dr=test_comb_dir)
```

Notice the different transformations such as rotation, flip, shift of width/height, zoom and shear (sort of distortion) is applied to the images. The drawback of this method is that images will need to be picked, processed in array to be fed to the data generator. So, we'll not be using this.

Lets look at <b>Option 2 - flow_from_directory</b>

In this, the different classes simply need to be in separate directories and Keras directly reads them from the main directory, applying categorical variables to classes automatically. It does mean that you need to have both, the training and testing images organized in subfolders. Although this could become cumbersome for large classes, we will be testing it here only on 2 classes to see the inner workings.

```python
# Now, try using FLOW from directory to see how the images get generated only fly by saving it in to output directory
# Note that we need to run a mini-CNN inorder to see it in action.
# I have manually put 10 sample images in the input directory for only 2 fruits Apple and Strawberry

path_keras_input_test = root_dir + 'ftd_TrainSet/'
path_keras_output = root_dir + 'ftd_Temp/'

# Define data generator values for training and validation set
train_datagen = ImageDataGenerator(
    rotation_range = 40,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    rescale = 1./255,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    validation_split=0.2               # Train-test split can be specified here.
    )

# Apply the flow from dir method on test objects
train_generator = train_datagen.flow_from_directory(
    path_keras_input_test,
    target_size=(100, 100),
    batch_size=32,
    class_mode='binary',
    save_to_dir=path_keras_output,
    seed=42,
    subset='training'
    )

validation_generator = train_datagen.flow_from_directory(
    path_keras_input_test,
    target_size=(100, 100),
    batch_size=32,
    class_mode='binary',
    save_to_dir=path_keras_output,
    seed=42,
    subset='validation'
    )

# Check if the train-test split is accurate
pd.Series(validation_generator.classes).value_counts()
```

```python
# Build a small model to test 
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),activation='relu',input_shape=(100,100,3), padding='same'))
model.add(Conv2D(32, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
%%time
model.fit_generator(
    train_generator,
    steps_per_epoch=3,
    epochs=10,
    validation_data=validation_generator,
    validation_steps=1)
```

Lets now try with <b> Option 3: flow_from_dataframe </b>. This seems to be cleaner in sense of applying scikit-learns train-test split before training the model. 

Whilst Keras provides both, transformation and train-test split options, it was taken up separately to validate each step in the process and also to manage is easily.

```python
# Load the train and test files

train = pd.read_csv(fruit_csv)
test = pd.read_csv(test_fruit_csv)

train.head()
```

```python
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img

df = train
df['id'] = df.id.astype(str)
df['label'] = df.label.astype(str)
df['id'] = df['id'] + '.jpg'

train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)
train_set.head()
```

```python
df_test = test
df_test['id'] = df_test.id.astype(str)
df_test['label'] = df_test.label.astype(str)
df_test['id'] = df_test['id'] + '.jpg'
df_test.head()
```

```python
train_set['label'].value_counts()
```

```python
# Define image generators

datagen=ImageDataGenerator(
    rotation_range = 45,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    shear_range = 0.3,
    zoom_range = 0.3,
    horizontal_flip = True,
    rescale=1./255)

datagentest=ImageDataGenerator(
    rescale=1./255)

train_generator=datagen.flow_from_dataframe(
    dataframe=train_set, 
    directory=comb_dir, 
    x_col="id", 
    y_col="label", 
    class_mode="categorical", 
    target_size=(100,100), 
    batch_size=32,
    shuffle=True)

valid_generator=datagentest.flow_from_dataframe(
    dataframe=test_set, 
    directory=comb_dir, 
    x_col="id", 
    y_col="label", 
    class_mode="categorical", 
    target_size=(100,100), 
    batch_size=32,
    shuffle=False)

test_generator=datagentest.flow_from_dataframe(
    dataframe=df_test, 
    directory=test_comb_dir, 
    x_col="id", 
    y_col="label", 
    class_mode="categorical", 
    target_size=(100,100), 
    batch_size=32,
    shuffle=False)
```

```python
# Best model (Other trainings happened in COLAB environment)
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),activation='relu',input_shape=(100,100,3), padding='same'))
model.add(Conv2D(32, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(5, activation='softmax'))
```

```python
from keras.callbacks import ModelCheckpoint
checkpoint = ModelCheckpoint('weights_cap2_3.hdf5', monitor='val_loss', save_best_only=True)
callbacks_list = [checkpoint]
    
```

```python
from keras import optimizers
model.compile(optimizers.rmsprop(lr=0.001),loss='categorical_crossentropy', metrics=["accuracy"])
```

```python
%%time
STEP_SIZE_TRAIN=train_generator.n//train_generator.batch_size
STEP_SIZE_VALID=valid_generator.n//valid_generator.batch_size
model.fit_generator(generator=train_generator,
                    steps_per_epoch=STEP_SIZE_TRAIN,
                    validation_data=valid_generator,
                    validation_steps=STEP_SIZE_VALID,
                    epochs=10,
                    callbacks=callbacks_list,
                    verbose=2)
```
